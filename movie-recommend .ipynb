```python
import pandas as pd
import ast

# --- 1. Define safe collapse function ---
def collapse(obj):
    L = []
    if isinstance(obj, str):
        try:
            obj = ast.literal_eval(obj)
        except:
            return ''
    if isinstance(obj, list):
        for i in obj:
            if isinstance(i, dict) and 'name' in i:
                L.append(i['name'].replace(" ", ""))
            elif isinstance(i, str):
                L.append(i.replace(" ", ""))
    return " ".join(L)

# --- 2. Apply to each metadata column ---
for col in ['cast', 'crew', 'genres', 'keywords']:
    movies[col] = movies[col].apply(collapse)

# --- 3. Convert overview to string safely ---
movies['overview'] = movies['overview'].astype(str)

# --- 4. Create clean tags column ---
movies['tags'] = movies['overview'] + " " + movies['genres'] + " " + movies['keywords'] + " " + movies['cast'] + " " + movies['crew']

# --- 5. Optional: clean spaces and lowercase ---
movies['tags'] = movies['tags'].apply(lambda x: x.lower().replace(",", "").replace(".", "").replace("  ", " "))

# --- 6. Keep final dataset ---
new = movies[['movie_id', 'title', 'tags']]
new.head()
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem (text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)    
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
        
```

```python
recommend('Batman')
```
